```python
# Initialize Faker
fake = faker.Faker()
fake_usa = faker.Faker('en_US')
fake_india = faker.Faker('en_IN')
fake_china = faker.Faker('zh_CN')
fake_canada = faker.Faker('en_CA')
fake_korea = faker.Faker('ko_KR')
fake_philippines = faker.Faker('en_PH')
fake_taiwan = faker.Faker('zh_TW')
fake_mexico = faker.Faker('es_MX')

# Constants
NUM_EMPLOYEES = 10000
DEPARTMENTS = {
    "Legal": 0.05,
    "Marketing": 0.10,
    "Administrative": 0.10,
    "Operations": 0.20,
    "Sales": 0.10,
    "Finance": 0.05,
    "I/T": 0.10,
    "Product": 0.20,
    "Human Resource": 0.10
}

SALARY_RANGES = {
    "Legal": (60000, 200000),
    "Marketing": (50000, 150000),
    "Administrative": (40000, 100000),
    "Operations": (50000, 150000),
    "Sales": (40000, 120000),
    "Finance": (50000, 150000),
    "I/T": (60000, 180000),
    "Product": (70000, 200000),
    "Human Resource": (50000, 120000)
}

H1B_COUNTRIES = {
    "India": 0.72,
    "China": 0.13,
    "Canada": 0.04,
    "South Korea": 0.03,
    "Philippines": 0.02,
    "Taiwan": 0.01,
    "Mexico": 0.01
}

LANGUAGES = ["Spanish", "Mandarin", "Hindi", "French", "German", "Korean", "Tagalog"]

# Generate Data
employees = []
ssn_set = set()

for _ in range(NUM_EMPLOYEES):
    # Generate social security number ensuring uniqueness
    ssn = fake_usa.ssn()
    while ssn in ssn_set:
        ssn = fake_usa.ssn()
    ssn_set.add(ssn)

    # Determine citizenship and name
    if random.random() < 0.40:
        country = random.choices(list(H1B_COUNTRIES.keys()), weights=H1B_COUNTRIES.values(), k=1)[0]
        if country == "India":
            first_name, last_name = fake_india.first_name(), fake_india.last_name()
        elif country == "China":
            first_name, last_name = fake_china.first_name(), fake_china.last_name()
        elif country == "Canada":
            first_name, last_name = fake_canada.first_name(), fake_canada.last_name()
        elif country == "South Korea":
            first_name, last_name = fake_korea.first_name(), fake_korea.last_name()
        elif country == "Philippines":
            first_name, last_name = fake_philippines.first_name(), fake_philippines.last_name()
        elif country == "Taiwan":
            first_name, last_name = fake_taiwan.first_name(), fake_taiwan.last_name()
        elif country == "Mexico":
            first_name, last_name = fake_mexico.first_name(), fake_mexico.last_name()
        citizenship = country
    else:
        first_name, last_name = fake_usa.first_name(), fake_usa.last_name()
        citizenship = "USA"

    # Assign gender to balance male and female employees
    gender = random.choice(["Male", "Female"])

    # Generate additional fields
    email = fake.email()
    phone = fake.phone_number()
    age = random.randint(22, 65)
    years_of_experience = random.randint(0, age - 21)  # assuming people start working at 21
    department = random.choices(list(DEPARTMENTS.keys()), weights=DEPARTMENTS.values(), k=1)[0]
    salary_range = SALARY_RANGES[department]
    salary = round(random.uniform(salary_range[0], salary_range[1]), 2)
    num_languages = random.choices([0, 1, 2], weights=[0.6, 0.3, 0.1], k=1)[0]
    languages_spoken = ", ".join(random.sample(LANGUAGES, num_languages))
    job_title = fake.job()

    employees.append({
        "First Name": first_name,
        "Last Name": last_name,
        "Email": email,
        "Phone": phone,
        "Gender": gender,
        "Age": age,
        "Job Title": job_title,
        "Years Of Experience": years_of_experience,
        "Salary": salary,
        "Department": department,
        "Languages": languages_spoken
    })
```

```python
# Convert to DataFrame and save to CSV
df = pd.DataFrame(employees)
df.to_csv('/content/employees.csv', index=False)
```

```python
import pandas as pd

# Constants
NUM_EMPLOYEES = 10000
DEPARTMENTS = {
    "Legal": 0.05,
    "Marketing": 0.10,
    "Administrative": 0.10,
    "Operations": 0.20,
    "Sales": 0.10,
    "Finance": 0.05,
    "I/T": 0.10,
    "Product": 0.20,
    "Human Resource": 0.10
}

# Calculate number of employees per department
department_counts = {dept: int(NUM_EMPLOYEES * proportion) for dept, proportion in DEPARTMENTS.items()}

# Calculate number of men and women per department
gender_distribution = {dept: {"Men": count // 2, "Women": count // 2} for dept, count in department_counts.items()}

# Display the gender distribution
gender_distribution_df = pd.DataFrame(gender_distribution).T
gender_distribution_df
```

```python
# Salary ranges for each department
SALARY_RANGES = {
    "Legal": (60000, 200000),
    "Marketing": (50000, 150000),
    "Administrative": (40000, 100000),
    "Operations": (50000, 150000),
    "Sales": (40000, 120000),
    "Finance": (50000, 150000),
    "I/T": (60000, 180000),
    "Product": (70000, 200000),
    "Human Resource": (50000, 120000)
}

# Calculate average salary for each department
average_salaries = {dept: (range_[0] + range_[1]) / 2 for dept, range_ in SALARY_RANGES.items()}

# Calculate total payroll for each department
total_payroll = {dept: count * average_salaries[dept] for dept, count in department_counts.items()}

# Calculate overall yearly payroll
yearly_payroll = sum(total_payroll.values())

yearly_payroll
```

```python
import numpy as np

# Define standard deviation parameters
sigma_age = 3  # Small variability for age
sigma_salary = df['Salary'].mean() * 0.1  # 10% of the mean salary
sigma_experience = 2  # Moderate variability for years of experience

# Perturb the attributes using Gaussian noise
df['Age'] = df['Age'] + np.random.normal(0, sigma_age, df['Age'].shape)
df['Salary'] = df['Salary'] + np.random.normal(0, sigma_salary, df['Salary'].shape)
df['Years Of Experience'] = df['Years Of Experience'] + np.random.normal(0, sigma_experience, df['Years Of Experience'].shape)

# Ensure the values remain realistic
df['Age'] = df['Age'].clip(lower=18)  # Assuming a minimum age of 18
df['Years Of Experience'] = df['Years Of Experience'].clip(lower=0)  # Experience cannot be negative
df['Salary'] = df['Salary'].clip(lower=0)  # Salary cannot be negative

# Display the perturbed sample
perturbed_sample = df.sample(20)
perturbed_sample
```
